```python
#Import dependencies
import keras
from data_handlers import *
import pandas as pd
```

```python
#load dataset
dataset = pd.read_csv("mushrooms.csv")
```

```python
#prepare dataset by encoding strings to numerical data using function from file data_handlers
encoded_dataset = prepare_dataset(dataset)
```

```python
#Spliting data to labels - y and features - x using function from file data_handlers
X, y = labels_vs_features(encoded_dataset)
```

```python
#split data into training and testing set
X_train, X_test, y_train, y_test = data_splitter(X, y)
```

```python
#We will need to reshape Y_train into [len(y), 1] becuase of the matrix multiplication
import numpy as np
y_train = np.reshape(y_train, (len(y_train), 1))
```

```python
#Importing keras Dense  (fully connected) layer and Sequential model
from keras.layers import Dense
from keras.models import Sequential
```

```python
#Creating model - fully connected network, no regularization techniques
model = Sequential()
model.add(Dense(50, input_shape = [22], activation='sigmoid'))
model.add(Dense(25, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error', metrics=['accuracy'])
```

```python
#Showing the structure of our neural network
model.summary()
```

```python
#setting final parameters for the network and it begins to train
number_of_iterations = 900
batch_size = 50
model.fit(X_train, y_train, batch_size=batch_size, epochs=number_of_iterations, verbose=2, validation_data=(X_test, y_test))
#Number of iterations run: 900
```

As you can see, MAX accuracy on this dataset, that I could get, in Keras with same NN used in other libraries is 99.94%
