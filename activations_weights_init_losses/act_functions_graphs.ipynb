```python
import numpy as np
import matplotlib.pyplot as plt
from activation_functions import  *
from losses import *
from weights_init import *
%matplotlib inline
```

```python
#Create data to plot using results from each activation function
x = np.arange(-10., 10., 0.2)
```

## Sigmoid

```python
plt.plot(x, sigmoid(x))
plt.title("Sigmoid graph")
plt.show()
```

```python
plt.plot(x, sigmoid(x, der=True))
plt.title("Sigmoid graph - derivation")
plt.show()
```

The Sigmoid function is defind in the range [0, 1]. This particularly makes the sigmoid function good in approximating functions that map into probability spaces.

Another usage of the Sigmoid activation function is on the output layer when we do binary classifcation (2 classes). And the score  of the sigmoid is used to determine what class is predicted.

Few notes to take care while using Sigmoid in DNN. You must take care of the way you init your weights, if they are init badly there is big chance that the gradient goes to 0, which is bad and this is called The Vanishing Gradient problem.

## Tanh

```python
plt.plot(x, tanh(x))
plt.title("Tanh")
plt.show()
```

```python
plt.plot(x, tanh(x, der=True))
plt.title("Tanh  - derivation")
plt.show()
```

The Tanh is defined on range [-1, 1].

This activation function is mostly use in RNNs.

## Relu

```python
plt.plot(x, relu(x))
plt.title("Relu")
plt.show()
```

```python
plt.plot(x, relu(x, der=True))
plt.title("Relu  - derivation")
plt.show()
```

The ReLU is much different from Sigmoid and Tanh above. In the process of training some gradients (x < 0) will be 0. Because of this property with good regularization technique, Dropout for example, we can train much deeper (larger) networks with same computational power. 

This activation fun. is mostly used in CNN networks to prevent vanishing gradient.

ProTip: Good weight init for using ReLU is he_init.

## Leaky relu

```python
plt.plot(x, leaky_relu(x))
plt.title("Leaky relu")
plt.show()
```

```python
plt.plot(x, leaky_relu(x, der=True))
plt.title("Leaky relu  - derivation")
plt.show()
```

Leaky Relu have similar properties but instead of thresholding vlaues less than zero to zero, Leaky Relu is giving small values (alpha) to those neurons. So by using Leaky Relu in your netowork every neuron will be 'fired' but some of them with really small value, alpha. 

In most caseds alpha = 0.001, but that is another hyperparameter that you will experiment with.
